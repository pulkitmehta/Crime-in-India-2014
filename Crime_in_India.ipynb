```python
data_districts=data.drop(columns=['States/UTs','Year'])
```

```python
data_districts.shape
```

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE()
```

```python
data_without_districts= data_districts.drop(columns=['District'])
```

```python
data_districts_tsne=tsne.fit_transform(data_without_districts)
```

```python
data_districts_tsne
```

```python
states_colors=[]
n=0
states_colors.append(0)
for i in range(1,802):
    if data['States/UTs'].values[i]==data['States/UTs'].values[i-1]:
        states_colors.append(n)
    else:
        n=n+1
        states_colors.append(n)
```

```python
len(states_colors)
```

```python
data['States/UTs'].value_counts()
```

```python
data_without_districts.iloc[0].sum()
```

```python
total_crime=[]
for i in range(802):
    total_crime.append(data_without_districts.iloc[i].values.sum())
```

```python
len(total_crime)
```

```python
total_crime=pd.DataFrame(total_crime, data_districts['District'],columns=['Total Crime'])
```

```python
total_crime.head(10)
```

```python
states_crime=(pd.DataFrame(total_crime.values.reshape(1,-1)[0],data['States/UTs'].values,))
```

```python
states_crime.tail(10)
```

```python
states_crime.index
```

```python
state_crime=[]
states=[]
states.append('Andhra Pradesh')
total=19480
for i in range(1,802):
    if states_crime.index[i]==states_crime.index[i-1]:
        total=total+total_crime.values.reshape(1,-1)[0][i]
    else:
        states.append(states_crime.index[i])
        state_crime.append(total)
        total=total_crime.values.reshape(1,-1)[0][i]
state_crime.append(1473+6615)
```

```python
len(state_crime)
```

```python
len(states)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
data_without_districts_pca=pca.fit_transform(data_without_districts)
```

```python
n=802

plt.figure(figsize=(18,6))
plt.title("States/UT Vs Districts")
plt.bar(data['States/UTs'].value_counts().keys(),data['States/UTs'].value_counts().values)
plt.xticks((range(data['States/UTs'].value_counts().values.shape[0])),data['States/UTs'].value_counts().keys()
          ,rotation=90)
plt.xlabel("States/UT")
plt.ylabel("No.of Districts")
plt.show()

plt.figure(figsize=(20,20))
plt.title("Pearson's Coorelation Coefficient")
plt.imshow(data.corr().values)
plt.colorbar()
plt.xticks((range(89)),data.columns,rotation=90)
plt.yticks((range(89)),data.columns)
plt.show()

plt.figure(figsize=(18,6))
plt.title("District vs Total Crime")
plt.bar(total_crime.index,total_crime.values.reshape(1,-1)[0])
plt.xticks(())
plt.xlabel("DISTRICTS")
plt.ylabel("Total Crime")
plt.show()

plt.figure(figsize=(10,10))
plt.title("State Vs Total Crime")
plt.barh(states,state_crime)
plt.ylabel("States")
plt.xlabel("Total Crime")
plt.show()

plt.figure(figsize=(20,5))
plt.title("HEATMAP of CRIME vs STATE")
plt.imshow(np.array(state_crime).reshape(1,-1))
plt.colorbar()
plt.xticks((range(36)),states,rotation=90)
plt.yticks(())
plt.xlabel("STATES")
plt.show()

plt.figure(figsize=(10,10))
plt.title("PCA Representation")
plt.scatter(data_without_districts_pca[:,0],data_without_districts_pca[:,1],c=states_colors)
plt.xlabel("Component_1")
plt.ylabel("Component_2")
plt.show()

plt.figure(figsize=(15,5))
plt.title("TSNE Representation")
plt.scatter(data_districts_tsne[:n,0],data_districts_tsne[:n,1], c=states_colors[:n])
plt.xlabel("Component_1")
plt.ylabel("Component_2")
plt.show()
```
